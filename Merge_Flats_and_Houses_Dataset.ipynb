# **Problem Statement** : Merging Datasets of Flats and Houses into a Unified Property Listing Dataset

# **Background**:
Real estate agencies often deal with diverse property types, including flats and houses, each with distinct attributes. To provide a comprehensive and user-friendly property listing platform, it is crucial to merge these distinct datasets into a single cohesive dataset. This will streamline property search and comparison for potential buyers or renters.

# **Objective**:
The aim of this project is to merge two different datasets containing information about flats and houses into a single dataset, creating a unified property listing dataset. This dataset will serve as a foundation for building an efficient and user-friendly property listing platform.

# **Datasets**:

*Dataset A*: Contains information about flats, including attributes such as property name, property type, society, price, price per square foot, area, number of bedrooms, bathrooms, balconies, additional rooms, address, floor number, facing direction, age of possession, nearby locations, description, furnishing details, features, and rating.

*Dataset B*: Contains information about houses, including similar attributes as Dataset A, such as property name, property type, society, price, price per square foot, area, number of bedrooms, bathrooms, balconies, additional rooms, address, floor number, facing direction, age of possession, nearby locations, description, furnishing details, features, and rating.

# **Deliverables**:
The successful completion of this project will result in a unified property listing dataset that combines the information from both flat and house datasets. The merged dataset should be clean, consistent, and ready for further analysis and integration into a property listing platform.

# **Impact**:
This merged dataset will enable real estate agencies and property listing platforms to offer a more comprehensive and user-friendly experience to potential buyers or renters. It will streamline property search, comparison, and decision-making, ultimately enhancing the efficiency of the real estate market.

# **Import Basic Libraries**

```python
import numpy as np
import pandas as pd
```

# **Import Datsets**

```python
flats = pd.read_csv('flats_cleaned.csv')
houses = pd.read_csv('houses_cleaned.csv')
```

# **Merge flats and Houses**

```python
df = pd.concat([flats,houses],ignore_index=True)
```

```python
df = df.sample(df.shape[0],ignore_index=True)
```
