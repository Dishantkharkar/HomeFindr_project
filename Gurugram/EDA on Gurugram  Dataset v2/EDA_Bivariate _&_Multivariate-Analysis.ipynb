# Import Basic Libararies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

# Import Dataset

```python
# for showing maximum columns which are present in dataset
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('gurgaon_properties_cleaned_v2.csv').drop_duplicates()
```

```python
df.head()
```

```python
df.columns
```

```python
# for categorical coliumn
cat_df= ['property_type','balcony','facing','agePossession']
```

```python
# Loop through each column for analysis
for column in cat_df:
    # bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df,x=df[column],y=df['price'])
    plt.title(f'Bivariate Analysis: {column} vs Price')
    plt.xlabel(column)
    plt.ylabel('Price')
    plt.tight_layout()
    plt.show()
```

## Insights
**Property_type vs price**
* The average price of a flat is higher than the average price of a house.
* There is a larger spread of prices for flats than for houses. This means that there are more expensive flats and more affordable flats than there are expensive houses and affordable houses.
* The median price of a flat is higher than the median price of a house. This means that half of the flats are more expensive than the median price, and half of the flats are less expensive than the median price. The same is true for houses.
* There are more flats in the data set than houses.
* Overall, the graph shows that flats are typically more expensive than houses

**balcony type and price**
* The scatterplot shows that there is a positive correlation between balcony type and price. This means that apartments with balconies tend to be more expensive than apartments without balconies. The correlation is not perfect, however, as there are some apartments with balconies that are less expensive than apartments without balconies.
* There are also some interesting patterns in the scatterplot. For example, the apartments with the largest balconies tend to be the most expensive. This is likely because apartments with large balconies offer more space and amenities, which are valued by buyers.
* Another interesting pattern is that there are a few apartments with no balconies that are more expensive than apartments with small balconies. This may be due to factors such as the location of the apartment, the size of the apartment, or the quality of the construction.
* Overall, the scatterplot shows that there is a positive correlation between balcony type and price. However, there are some exceptions to this rule. 

**facing and price**
* The scatterplot shows that there is no clear correlation between facing and price. This means that there is no consistent relationship between the direction a house faces and its price. There are houses facing all directions that are both expensive and affordable.
* There are a few interesting patterns in the scatterplot. For example, there are a few houses facing south that are more expensive than houses facing north. This may be due to factors such as the location of the house, the size of the house, or the quality of the construction.
* Another interesting pattern is that there are a few houses facing north-west that are less expensive than houses facing south-west. This may be due to the fact that north-west facing houses are more likely to be exposed to wind and rain.
* Overall, the scatterplot shows that there is no clear correlation between facing and price. However, there are some exceptions to this rule. Further analysis of the data would be needed to understand these exceptions.

**agepossession vs price**
* Older properties tend to be less expensive than newer properties.
* The correlation between age possession and price is not perfect.
* New properties have the highest average price.
* Moderately old properties have the lowest average price.


```python
for column in cat_df:
    # box plot
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df,x=df[column],y=df['price'])
    plt.title(f'Bivariate Analysis: {column} vs Price')
    plt.xlabel(column)
    plt.ylabel('Price')
    plt.tight_layout()
    plt.show()
```
